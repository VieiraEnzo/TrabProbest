# Leitura dos dados

```python
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
smart = pd.read_csv('dataset_smart-tv.csv')
chromecast = pd.read_csv('dataset_chromecast.csv')
```

```python
smart['upLog'] = smart.bytes_up.apply(lambda x: x if x == 0 else math.log(x))
smart['downLog'] = smart.bytes_down.apply(lambda x: x if x == 0 else math.log(x))

chromecast['upLog'] = chromecast.bytes_up.apply(lambda x: x if x == 0 else math.log(x))
chromecast['downLog'] = chromecast.bytes_down.apply(lambda x: x if x == 0 else math.log(x))
```

```python
chromecast.head()
```

```python
DownloadColor = 'red'
UploadColor = 'skyblue'
```

## Média, Variancia e Desvio padrão

```python
#Para smart-tv

meanSmartUp = smart.bytes_up.mean()
varianceSmartUp = smart.bytes_up.var()
desvSmartUp = smart.bytes_up.std()

meanSmartDown = smart.bytes_down.mean()
varianceSmartDown = smart.bytes_down.var()
desvSmartDown = smart.bytes_down.std()

print("<Smart TV> ")
print("Média Up: ",meanSmartUp)
print("Média Down: ",meanSmartDown)
print("Var Up: ", varianceSmartUp)
print("Var Down: ",varianceSmartDown)
print("Desv Up: ", desvSmartUp)
print("Desv Down: ",desvSmartDown)

#Para chromecast

meanChromeUp = chromecast.bytes_up.mean()
varianceChromeUp = chromecast.bytes_up.var()
desvChromeUp = chromecast.bytes_up.std()

meanChromeDown = chromecast.bytes_down.mean()
varianceChromeDown = chromecast.bytes_down.var()
desvChromeDown = chromecast.bytes_down.std()

print("<Chromecast> ")
print("Média Up: ",meanChromeUp)
print("Média Down: ",meanChromeDown)
print("Var Up: ", varianceChromeUp)
print("Var Down: ",varianceChromeDown)
print("Desv Up: ", desvChromeUp)
print("Desv Down: ",desvChromeDown)
```

## função de distribuição empírica


### Smart

```python
def FuncDistEmp(lista): 
    lista.sort()
    return [(i+1)/len(lista) for i in range(len(lista))]

x1 = smart['downLog'].to_list()
x1.sort()

plt.plot(x1, FuncDistEmp(smart['downLog'].to_list()), marker='.', linestyle='none', color='red')

plt.title('ECDF download')
plt.xlabel('Valores dos dados')
plt.ylabel('Proporção cumulativa')

plt.show()

###########################################################################

x1 = smart['upLog'].to_list()
x1.sort()

plt.plot(x1, FuncDistEmp(smart['upLog'].to_list()), marker='.', linestyle='none', color='skyblue')

plt.title('ECDF upload')
plt.xlabel('Valores dos dados')
plt.ylabel('Proporção cumulativa')

plt.show()
```

### Chromecast

```python
x1 = chromecast['downLog'].to_list()
x1.sort()

plt.plot(x1, FuncDistEmp(chromecast['downLog'].to_list()), marker='.', linestyle='none', color='red')

plt.title('ECDF download')
plt.xlabel('Valores dos dados')
plt.ylabel('Proporção cumulativa')

plt.show()

###########################################################################

x1 = chromecast['upLog'].to_list()
x1.sort()

plt.plot(x1, FuncDistEmp(chromecast['upLog'].to_list()), marker='.', linestyle='none', color='skyblue')

plt.title('ECDF upload')
plt.xlabel('Valores dos dados')
plt.ylabel('Proporção cumulativa')

plt.show()
```

# Estatística por horario

Re-organizando os dados

```python
smart['date_hour'] = pd.to_datetime(smart['date_hour'])
smart = smart.set_index('date_hour')

chromecast['date_hour'] = pd.to_datetime(chromecast['date_hour'])
chromecast = chromecast.set_index('date_hour')
```

```python
groupedSmart = smart.groupby(smart.index.hour)

meanSmart = groupedSmart.mean()
varianceSmart = groupedSmart.var()
std_devSmart = groupedSmart.std()

############################################

groupedChrome = chromecast.groupby(chromecast.index.hour)

meanChrome = groupedChrome.mean()
varianceChrome = groupedChrome.var()
std_devChrome = groupedChrome.std()

```
